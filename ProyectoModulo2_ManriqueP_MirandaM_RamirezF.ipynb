## <font color = "green" > **Competencia**: 

Se refiere a la acción recíproca entre dos organismos que pretenden apoderarse de la misma cosa. En ecología, la competencia entre especies es toda acción recíproca entre poblaciones de diversas especies, que afecta adversamente su desarrollo o supervivencia.


 ###  <font color = "green" > **Norma ecológica: **</font>
 Afirma que dos especies no pueden ocupar un mismo nicho ecológico en el mismo hábitat al mismo tiempo
 
Es una proposición que declara que dos especies en competencia biológica por los mismos recursos no puede coexistir en forma estable si los demás factores ecológicos permanecen constantes. Uno de los competidores siempre dominará al otro, llevándolo a la extinción o a una modificación evolutiva o de comportamiento hacia otro nicho ecológico.

## <font color = "green" > **Características del principio de exclusión:** 

* La interacción competitiva puede incluir espacio, alimento o nutrientes, luz, materiales de desecho, susceptibilidad a los carnívoros, enfermedad y muchos otros tipos de interacciones mutuas.

* Los resultados de la competencia han sido de gran interés para los biólogos evolucionistas y ha sido ampliamente estudiado como uno de los mecanismos de selección natural. 

* Los recursos están en cantidades limitadas: recursos en poca cantidad causan competencia intra e interespecífica. 

* Los coeficientes de competencia y las capacidades de carga son constantes: no cambian con la densidad, ni el tiempo o el espacio.

#  <span style="color:blue">**1.3 Modelo matemático**</span>


### <span style="color:green">**Comportamiento de la primera levadura **</span>
\begin{align}
\frac{dx}{dt}&=a_1x - a_2x^2 - a_3xy
\end{align}

Donde $a1x - a2x^2$ describe su comportamiento cuando se encuentra sola, mientras que el término $-a3xy$ se refiere a la competencia que producirá con la segunda levadura al entrar en contacto

### <span style="color:green">**Comportamiento de la segunda levadura **</span>

\begin{align}
\frac{dy}{dt}&=b_1y - b_2y^2 - b3xy
\end{align}

Donde $b1y - b2y^2$ describe su comportamiento cuando se encuentra sola, mientras que el término $-b3xy$ se refiere a la competencia que producirá con la segunda levadura al entrar en contacto

### <span style="color:green">** Parámetros del comportamiento de las levaduras **</span> 

* $a_1 = 0.2586$                           
* $a_2 = 0.02030$                          
* $a_ 3 = 0.05711$                         
* $b_1 = 0.0574$
* $b_2 = 0.00976$
* $b_3 = 0.00480$

Una vez definidas los parámetros que se adaptan de mejor manera al comportamiento de las levaduras estudiadas, el modelo queda de la siguiente manera:

\begin{align}
\frac{dx}{dt}&=0.2586x - 0.02030x^2 - 0.05711xy
\end{align}

\begin{align}
\frac{dy}{dt}&=0.0574y - 0.00976y^2 - 0.00480xy
\end{align}


#  <span style="color:blue">**1.4 Simulaciones**</span>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def ecuaciones(x,t, a1, a2, a3, b1, b2, b3):
    x1 = x[0]
    x2 = x[1]
    return [a1*x1-a2*x1**2-a3*x1*x2, b1*x2-b2*x2**2-b3*x2*x1]

x0 = [.4184, .6315]

t = np.linspace(0, 1e3, 10000)

a1 = 0.2586
a2 = 0.0203
a3 = 0.05711
b1 = 0.0574
b2 = 0.00976
b3 = 0.0048
```

```python
x = odeint(ecuaciones, x0, t, args=(a1, a2, a3, b1, b2, b3))
x1 = x[:,0]
x2 = x[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, x1, 'r', lw=2, label='$x_1(t)$: Levadura 1')
plt.plot(t, x2, 'b', lw=2, label='$x_2(t)$: Levadura 2')
plt.title('Comportamiento de las levaduras en el tiempo')
plt.grid()
plt.legend(loc='best')
plt.xlabel('t [horas]')
plt.ylabel('Volumen poblacional')
plt.show()
```

* En las primeras horas La levadura 1 y 2  aumentan su volumen poblacional, siendo la primera la que tiene mayor volumen de crecimiento
* El crecimiento poblacional de la levadura 1 llega a su máximo y comienza a decaer hasta desaparecer.
* El crecimiento poblacional de la levadura 2 llega a su máximo y se mantiene constante, lo que le permite coexistir
* La levadura 2 coexiste en el hábitat


```python
plt.figure(figsize=(6,6))
plt.plot(x1, x2, lw=2, label='trayectoria')
plt.grid()
plt.plot(x1[1], x2[1], 'o', ms=5, label='condición inicial')
plt.plot(x1[-1], x2[-1], 'o', ms=5, label='condición final')
plt.title('Diagrame fase')
plt.xlabel('$x_1$ [Levadura 1]')
plt.ylabel('$x_2$ [Levadura 2]')
plt.legend(loc='best')
plt.show()
```
